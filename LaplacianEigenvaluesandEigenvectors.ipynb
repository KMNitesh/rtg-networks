```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
def create_weight(edgelist,n):
    W = np.zeros((n,n))
    for i,j in edgelist:
        W[i,j] = 1
        W[j,i] = 1
    return W
def create_Lap(W):
    D = np.diag(W.sum(axis=0))
    L = D - W
    return L
```

```python
#path graph
n = 11
a = nx.path_graph(n,create_using=None)
edgelist_a = a.edges()
W = create_weight(edgelist_a,n)
L = create_Lap(W)
(e_vals, e_vecs) = np.linalg.eig(L)
eig_vals_sorted = np.sort(e_vals)
eig_vecs_sorted = e_vecs[:, e_vals.argsort()]
nx.draw(a)
plt.show()
```

```python
#plot eig values
plt.plot(eig_vals_sorted)
plt.show()
```

```python
#plot second and third eigenvector
index=np.argsort(e_vecs[1])
v_2 = np.sort(e_vecs[1])
v_3 = list(map(lambda x: e_vecs[2][x], index))
plt.plot(v_2, v_3)
plt.show()
```
