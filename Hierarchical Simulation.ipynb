```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

np.random.seed(7412)  # for repeatability of this simulation
a = np.random.multivariate_normal([0, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 5], [[3, 1], [1, 4]], size=[100,])
c = np.random.multivariate_normal([3, 10], [[3, 1], [1, 4]], size=[100,])
d = np.random.multivariate_normal([3, 15], [[3, 1], [1, 4]], size=[100,])
e = np.random.multivariate_normal([6, 25], [[3, 1], [1, 4]], size=[100,])
f = np.random.multivariate_normal([6, 30], [[3, 1], [1, 4]], size=[100,])
g = np.random.multivariate_normal([9, 35], [[3, 1], [1, 4]], size=[100,])
h = np.random.multivariate_normal([9, 40], [[3, 1], [1, 4]], size=[100,])

L = np.concatenate((a, b, c, d))
M = np.concatenate((e, f, g, h))

X = np.concatenate((L, M))

plt.scatter(X[:,0], X[:,1])
plt.show()

plt.scatter(L[:,0], L[:,1],color='yellow')
plt.scatter(M[:,0], M[:,1],color='blue')
plt.show()
```

```python
N = np.concatenate((a, b))
O = np.concatenate((c, d))
P = np.concatenate((e, f))
Q = np.concatenate((g, h))

Y = np.concatenate((N, O))
Z = np.concatenate((P, Q))

plt.scatter(N[:,0], N[:,1],color='green')
plt.scatter(O[:,0], O[:,1],color='red')
plt.scatter(P[:,0], P[:,1],color='orange')
plt.scatter(Q[:,0], Q[:,1],color='purple')
plt.show()
```
